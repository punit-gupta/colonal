```python
def Fun(variables_values = [0, 0]):
  list1 = [0] * data['vmcount'][0]
  val = 0.0
  solution=variables_values
  #print(solution)
  #print(int(round(solution[0])))
  for i in range(solution.size):
    solution[i]=round(solution[i])
    if(round(solution[i])!=0.0):
      list1[int(round(solution[i]))-1]=list1[int(round(solution[i]))-1] + data['task'][i]/data['VM'][int(round(solution[i]))-1]
    else:
      list1[int(round(solution[i]))]=list1[int(round(solution[i]))] + data['task'][i]/data['VM'][int(round(solution[i]))]
  #print(solution)
  return max(list1)

start_time = time.time()
```

```python
# Chicken-SO - Parameters
parameters = {
    'size': 10,
    'min_values': [0 for i in range(size)],
    'max_values': [vmcount-1 for i in range(size)],
    'iterations': 50,
    'clone_factor': 0.1,
    'num_rand': 2,
	  'verbose': True,
	  'start_init': None,
	  'target_value': None
}
```

```python
# Chicken-SO - Algorithm
chicken_so = clonal_selection_algorithm(target_function = Fun, **parameters)
```
